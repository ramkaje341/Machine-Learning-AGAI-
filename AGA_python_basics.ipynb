```python
def computePay(hours, pay_rate):
  if hours <= 40:
    total_pay = hours * pay_rate
  else:
    regular_pay = 40 * pay_rate
    overtime_hours = hours - 40
    overtime_pay = overtime_hours * (pay_rate * 1.5)
    total_pay = regular_pay + overtime_pay
  return total_pay


try:
  hours_worked = float(input("Enter hours: "))
  pay_rate = float(input("Enter pay rate: "))

  total_pay = computePay(hours_worked, pay_rate)
  print(f"Total pay: {total_pay}")

except ValueError:
  print("Invalid input. Please enter numerical values for hours and pay rate.")
```

```python
def bubble_sort(arr):
    n = len(arr)

    for i in range(n):
        for j in range(0, n - i - 1):
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
    return arr

arr = [89 , 14 , 21, 34 , 55 , 3]
sorted_arr = bubble_sort(arr)
print("Sorted array:", sorted_arr)
```

Insertion sort in arrays

```python
def insertion_sort(arr):
    n = len(arr)
    for i in range(1, n):
        key = arr[i]

        j = i - 1
        while j >= 0 and key < arr[j]:
            arr[j + 1] = arr[j]
            j -= 1
        arr[j + 1] = key

list = [89 , 14 , 21, 34 , 55 , 3]
insertion_sort(list)
print("Sorted array is:",list)
```

Quick sort in list


```python
def quick_sort(arr):
    if len(arr) <= 1:
        return arr
    else:
        pivot = arr[len(arr) // 2]
        less = [x for x in arr if x < pivot]
        equal = [x for x in arr if x == pivot]
        greater = [x for x in arr if x > pivot]
        return quick_sort(less) + equal + quick_sort(greater)


list = [89 , 14 , 21, 34 , 55 , 3]
sorted_data = quick_sort(list)
print("Sorted Array:", sorted_data)
```

17. Let x = np.array([[1,2],[3,4]])
    Write the code to perform the following
	(a) Compute sum of all elements of x.
	(b) Compute sum of each column of x.
	(c) Compute sum of each row of x.


```python
import numpy as np

x = np.array([[1, 2],
              [3, 4]])


total_sum = np.sum(x)

column_sum = np.sum(x, axis=0)

row_sum = np.sum(x, axis=1)

print("Sum of all elements:", total_sum)
print("Sum of each column:", column_sum)
print("Sum of each row:", row_sum)
```
